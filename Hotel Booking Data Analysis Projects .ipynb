Now we have to calculate the MEAN considering the reservation date

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

now we will create visialization for the average daily rate for both city as well as resort hotel

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
# Plot the average daily rate for Resort Hotel
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
# Plot the average daily rate for City Hotel
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend()
plt.show()
```

Now we have to find out in which month the reservations were highest and
in which month the cancellations were highest

```python
# Extract the month from the 'reservation_status_date' column and assign it to a new column 'month'
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x='month', hue = 'is_canceled', data=df)
ax1.set_ylabel('Number of Reservations')
plt.legend(['not cancelled', 'cancelled'])
```

Now we just need to look at the data on canceled reservations to know if price has any impact on the cancellation rate.

```python

# Filter the DataFrame to include only canceled bookings
cancelled_bookings = df[df['is_canceled'] == 1]

# Group by month and calculate the sum of average daily rate (ADR) for canceled bookings
cancelled_adr_by_month = cancelled_bookings.groupby('month')[['adr']].sum().reset_index()

# Create a bar plot
sns.barplot(x='month', y='adr', data=cancelled_adr_by_month)

# Set x-axis label
plt.xlabel('Month')

# Set y-axis label
plt.ylabel('Total ADR for Canceled Bookings')

# Set title
plt.title('Total ADR for Canceled Bookings by Month')
```

Now we have to visualize the data as per top 10 countries

```python

# Filter the DataFrame to include only canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Count the number of canceled bookings for each country and select the top 5 countries
top_5_country = cancelled_data['country'].value_counts()[:5]

# Create a figure with specified size
plt.figure(figsize=(8, 8))

# Set the title for the pie chart
plt.title('Top 5 countries with reservations canceled')

# Create a pie chart with the top 5 countries and their corresponding counts
plt.pie(top_5_country, autopct='%.2f', labels=top_5_country.index)

# Show the pie chart
plt.show()
```
